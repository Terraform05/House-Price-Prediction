```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/train.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Histograms of the numeric features

numeric_df = df.select_dtypes(include=[np.number])
a = int(np.ceil(np.sqrt(numeric_df.shape[1])))
b = int(np.ceil(numeric_df.shape[1] / a))

pl, ax = plt.subplots(a, b, figsize=(20, 20))
for i, col in enumerate(numeric_df.columns):
	c = numeric_df[col]
	ax[i // b, i % b].hist(c, bins=20)
	ax[i // b, i % b].set_title(col)

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

pl, ax = plt.subplots(a, b, figsize=(20, 20))
for i, col in enumerate(numeric_df.columns):
	lm = LinearRegression()
	# Drop samples with missing values
	X = numeric_df[col].dropna().values.reshape(-1, 1)
	y = df["SalePrice"].loc[numeric_df[col].dropna().index]
	lm.fit(X, y)
	ax[i // b, i % b].scatter(X, y)
	ax[i // b, i % b].plot(X, lm.predict(X), color="red")
	ax[i // b, i % b].set_title(col)

plt.tight_layout()
plt.show()
```
